Feature Selection

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []

# Iterate over columns in the DataFrame
for col in df.columns:
    # Check if column is numeric and not 'price'
    if df[col].dtype != 'object' and col != 'price':
        col_list.append(col)

# Select columns based on col_list
x = df[col_list]

# Initialize DataFrame for VIF data
vif_data = pd.DataFrame()

# Populate 'Feature' column
vif_data['Feature'] = x.columns

# Calculate VIF for each feature
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]

# Display DataFrame with VIF data
print(vif_data)
```

Linear Regression Algorithm

```python
x = df.drop(columns='price')
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=24)
```

```python
from sklearn.preprocessing import StandardScaler
sc  = StandardScaler()

x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
differenc= pd.DataFrame(np.c_[y_test,y_pred], columns=['Actual ', 'Predicted'])
```

```python
differenc
```

```python
sns.distplot(y_pred,label='Prediction')
sns.displot(y_test, label='Actual')
plt.legend()
```

```python
from sklearn.metrics import r2_score, mean_absolute_percentage_error
from sklearn import metrics
```

```python
r2_score(y_pred,y_test)
```
